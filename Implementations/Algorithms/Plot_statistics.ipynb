```python
import matplotlib.pyplot as plt
from datetime import datetime
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
workdir = "~/mestrado/MAC4788/ReinforcementLearning-knowledgeBase/Implementations/Algorithms/data/"

def load_df(filename, experiment_name=""):
    file_path = workdir+filename
    index = ["episode","total_reward","window_avg","window_min","window_max","window_std_dev"]
    df = pd.read_csv(file_path, names=index)
    df['experiment'] = filename.split('.')[0] if experiment_name == "" else experiment_name
    return df

# [(experiments_results_filename, experiment_description)]
experiment_data = [("pong_5981_R2G_frame_diff.csv", "Reward 2 go and frame diff"),
             ("pong_10000_R2G_frame_sum.csv", "Reward 2 go and frame sum"),
             ("pong_6121_R2G_no_frame_diff.csv", "Reward 2 go")]

data = [load_df(f, e) for f, e in experiment_data]

line_limit = 10000
for d in data:
    d_size = d.shape[0]
    line_limit = d_size if d_size < line_limit else line_limit
```

```python
sns.lineplot(
    # exps eh uma lista de algum objeto generico representando os resultados do experimento
    data=pd.concat([exp.head(line_limit) for exp in data], ignore_index=True, sort=False),
    x="episode",
    y="total_reward",
    hue="experiment", # curvas de cores diferentes para cada algoritmo ou variacao
#     estimator="mean", # as curvas sao a media das rodadas de cada algoritmo
)
```
